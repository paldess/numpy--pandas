### Задание 3

```python
x = np.linspace(-3, 3, 51)
```

```python
y1 = x**2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
ax_1, ax_2, ax_3, ax_4 = ax.flatten()
ax_1.plot(x, y1)
ax_1.set_title('график y1')
ax_1.set_xlim(-5, 5)

ax_2.plot(x, y2)
ax_1.set_title('график y2')

ax_3.plot(x, y3)
ax_1.set_title('график y3')

ax_4.plot(x, y4)
ax_1.set_title('график y4')

fig.set_size_inches(8, 6)
plt.subplots_adjust(wspace=0.3, hspace=0.3)

plt.show()
```
