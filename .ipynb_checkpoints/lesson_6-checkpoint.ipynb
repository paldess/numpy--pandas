### Задание 1

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_boston
```

```python
data1 = load_boston()
data1.keys()
```

```python
data = pd.DataFrame(data1['data'], columns=data1['feature_names'])
data.head()
```

```python
target = pd.DataFrame(data1['target'], columns=['price'])
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(data, target, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(x_train, y_train)
y_pred = lr.predict(x_test)
y_pred
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

### Задание 4

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV, train_test_split 
```

```python
data = pd.read_csv('creditcard.csv')
data = pd.DataFrame(data)
data
```

```python
data.value_counts(normalize=True)
```

```python
pd.options.display.max_columns = 100
data.info()
```

```python
data.head(10)
```

```python
class_series = pd.Series(data['Class'])
class_series
```

```python
x = data
x.drop(columns='Class', inplace=True)
x
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, class_series, test_size=0.3, random_state=100, stratify=class_series)
```

```python
x_train.info()
```

```python
y_train.head()
```

```python
parameters = [{'n_estimators': [10, 15],
'max_features': np.arange(3, 5),
'max_depth': np.arange(4, 7)}]
```

```python
from sklearn.model_selection import GridSearchCV
avg = GridSearchCV(estimator=RandomForestClassifier(random_state=100),
param_grid=parameters,
scoring='roc_auc',
cv=3)
avg.fit(x_train, y_train)
```

```python
y_pred = avg.predict(x_test)
```

```python
y_pred == y_test
```

```python
from sklearn.metrics import roc_auc_score
```

```python
roc_auc_score(y_pred, y_test)
```

```python
y_pred_train = avg.predict(x_train)
```

```python
roc_auc_score(y_pred_train, y_train)
```

```python
plt.figure(figsize=(15, 5))
# plt.ylim(0.98, 1.1)
plt.plot(y_pred, color='red')
plt.plot(y_pred_train, color='green')
plt.show()
```
