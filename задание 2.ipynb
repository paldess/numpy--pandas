```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
from sklearn.datasets import load_boston
from sklearn.preprocessing import StandardScaler
from matplotlib import pyplot as plt
from sklearn.manifold import TSNE
```

```python
boston = load_boston()
boston.keys()
```

```python
data = pd.DataFrame(boston['data'], columns=boston['feature_names'])
target = pd.DataFrame(boston['target'], columns=['price'])
```

```python
x_train, x_test, y_train, y_test = train_test_split(data, target, random_state=42, test_size=20)
x_test.head()
```

```python
scaler = StandardScaler()
x_train_scaler = scaler.fit_transform(x_train)
x_train_scaler = pd.DataFrame(x_train_scaler, columns=boston['feature_names'])
x_train_scaler
```

```python
tsne = TSNE(random_state=42, learning_rate=250, n_components=2)
```

```python
x_train_tsne = tsne.fit_transform(x_train_scaler)
x_train_tsne
```

```python
plt.scatter(x_train_tsne[:, 0], x_train_tsne[:, 1])
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, random_state=42, max_iter=100)
label_train = kmeans.fit_predict(x_train_scaler)
pd.value_counts(label_train)
```

```python
labels = kmeans.labels_
```

```python
kmeans.cluster_centers_
```
